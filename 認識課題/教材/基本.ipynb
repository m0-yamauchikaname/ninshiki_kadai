
# Pythonで学ぶ画像処理の基礎

このノートブックでは、Pythonを使用して画像処理を学ぶ方法を紹介します。Pythonの画像処理ライブラリである **OpenCV** を中心に、以下の内容を学習します：

1. 画像の読み込みと表示
2. グレースケール変換
3. エッジ検出
4. 画像の保存

必要なライブラリをインストールして、次に進みましょう！


```python

# 必要なライブラリをインストールします（初回のみ必要）
# !pip install opencv-python matplotlib

# ライブラリのインポート
import cv2
import matplotlib.pyplot as plt
```

```python
#############
# opencvバージョン
#############
# sleepしたいのでインポート
import time

# 画像の読み込み
image_path = "Lenna.jpg"  # 任意の画像ファイルを指定してください
image_cv = cv2.imread(image_path)

# 画像の表示
cv2.imshow("Original Image", image_cv)
# キー入力待ち
cv2.waitKey(1)

# 3秒待機
time.sleep(3)

# ウィンドウを閉じる
cv2.destroyAllWindows()
```

```python
####################
# matplotlibバージョン
####################

# 画像の読み込み
image_path = "Lenna.jpg"  # 任意の画像ファイルを指定してください
image = cv2.imread(image_path)

# 画像の表示（OpenCVはBGR形式なので、RGBに変換）
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python

# グレースケール変換
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python

# Cannyエッジ検出
edges = cv2.Canny(gray_image, 100, 200)

plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

```python

# 処理結果を保存
output_path = "output_edges.jpg"
cv2.imwrite(output_path, edges)
print(f"エッジ検出結果を保存しました: {output_path}")
```
